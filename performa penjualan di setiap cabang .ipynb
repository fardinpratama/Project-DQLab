Perbandingan performa tiap cabang dari 5 kota dari segi order size, customer count, product count, brand count, dan GMV dalam basis bulanan.

Tujuan : Automation untuk pembuatan grafik dari measurement yang dibutuhkan 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

data - data yang digunakan:
- csv (data dari bulan January - March)       --> https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/LO4/retail_data_from_1_until_3.csv
- csv (data dari bulan April - June)              --> https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/LO4/retail_data_from_4_until_6.csv
- csv (data dari bulan July - September)      --> https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/LO4/retail_data_from_7_until_9.csv
- csv (data dari bulan October - December)  --> https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/LO4/retail_data_from_10_until_12.csv

*) data yang digunakan sudah bersih

```python
# [1]. Load masing-masing data dengan pandas
retail_data1 = pd.read_csv('https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/LO4/retail_data_from_1_until_3.csv')
retail_data2 = pd.read_csv('https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/LO4/retail_data_from_4_until_6.csv')
retail_data3 = pd.read_csv('https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/LO4/retail_data_from_7_until_9.csv')
retail_data4 = pd.read_csv('https://dqlab-dataset.s3-ap-southeast-1.amazonaws.com/LO4/retail_data_from_10_until_12.csv')
```

semua data frame memilik list kolom yang sama, sehingga kita bisa langsung menggabungkannya.

```python
retail_table = pd.concat([retail_data1, retail_data2, retail_data3, retail_data4])
print('Jumlah baris:', retail_table.shape[0])
```

Melakukan transformasi data

```python
#memasitikan data yang memiliki item_price < 0 atau total_price < 0
cek = retail_table.loc[(retail_table['item_price'] < 0) | (retail_table['total_price'] < 0)]
cek
```

terlihat dua data ini memiliki item_price dan total_price memilki nilai yang tidak masuk akal (nilai minus) seghingga data-data ini dapat dibuang

```python
if cek.shape[0] != 0:
    retail_table = retail_table.loc[(retail_table['item_price'] > 0) & (retail_table['total_price'] > 0)]
```

 Cek apakah masih ada order_id yang bernilai undefined dan delete row tersebut

```python
cek = retail_table.loc[retail_table['order_id'] == 'undefined']
cek
```

terlihat masih terdapat satu data yang memiliki order_id bernilai undifined

```python
if cek.shape[0] != 0:
    retail_table = retail_table.loc[retail_table['order_id'] != 'undefined']
```

Transform order_id menjadi int64 dan order_date menjadi datetime Pandas

```python
retail_table['order_id'] = retail_table['order_id'].astype('int64')
retail_table['order_id'].dtypes
```

```python
retail_table['order_date'] = pd.to_datetime(retail_table['order_date'])
retail_table['order_date'].dtypes
```

```python
#cek kembali data
retail_table.describe()
```

 Filter province yang hanya termasuk 5 provinsi besar di Jawa (DKI Jakarta, Jawa Barat, Jawa Tengah, Jawa Timur, dan Yogyakarta)

```python
java = ['DKI Jakarta','Jawa Barat','Jawa Tengah','Jawa Timur','Yogyakarta']
retail_table = retail_table.loc[retail_table['province'].isin(java)]
```

```python
#Memastikan filter provinsi telah berhasil
print(retail_table.province.unique())
```

Mengelompokkan data berdasarkan order_date dan province yang sudah difilter dan menghitung order unique count, customer unique count, product unique count, brand unique count, dan GMV (Gross Merchandise Volume = total_price untuk semua penjualan)

```python
groupby_city_province = retail_table.groupby(['order_date','province']).agg({
   'order_id': 'nunique',
   'customer_id': 'nunique',
   'product_id': 'nunique',
   'brand': 'nunique',
   'total_price': sum
})
```

mengubah nama kolomnya menjadi 'order','customer','product','brand','GMV'

```python
groupby_city_province.columns = ['order','customer','product','brand','GMV']
```

```python
groupby_city_province.head(10)
```

Unstack untuk mendapatkan order_date di bagian baris dan province di bagian column

```python
unstack_city_province = groupby_city_province.unstack(level='province').fillna(0)
unstack_city_province.head()
```

Buat grafik dari masing-masing pengukuran dengan terlebih dahulu melakukan slicing dan resampling untuk dilakukan perhitungan rata-rata bulanan pada masing-masing pengungukuran.

```python
# Create figure canvas dan axes for 5 line plots
fig, axes = plt.subplots(5, 1, figsize=(8, 18))

# Slicing index
idx = pd.IndexSlice['order','customer','product','brand','GMV']
for i, measurement in enumerate(groupby_city_province.columns):
    # Slicing data untuk masing-masing measurement
    by_measurement = unstack_city_province.loc[:,idx[i]]
    # Lakukan resampling pada data tersebut untuk dilakukan perhitungan rata-rata bulanan 
    by_measurement_monthly_mean = by_measurement.resample('M').mean()
    # Plot by_measurement_monthly_mean
    by_measurement_monthly_mean.plot(
        title = 'Average Daily ' + measurement + ' Size in Month View for all Province',
        ax = axes[i]
    )
    axes[i].set_ylabel('avg ' + measurement + ' size')
    axes[i].set_xlabel('month')

# Adjust the layout and show the plot
plt.tight_layout()
plt.show()
```

Dari grafik di atas terlihat rata-rata ukuran performa tertinggi berada di provinsi DKI Jakarta, kemudian disusul oleh provinsi Jawa barat, tiga provinsi sisanya berada di bawah dalam angka yang rata-rata sama.

Sebagian besar performa dari cabang-cabang di 5 provinsi mengalami kenaikan namun sempat terjadi penurunan di bulan september terutama pada cabang di provinsi Jawa Timur
